## 01 맷플롯립
- 데이터 시각화는 데이터 분석 결과를 쉽게 이해할 수 있도록 시각적으로 표현하고 전달되는 과정을 의미.

### 맷플롯립
- 매트랩에서 사용하는 다양한 기능을 파이썬에서 그대로 사용할 수 있도록 해주는 시각화 도구이다.
- 엑셀에서 사용하는 정형화된 차트나 그래프를 작성할 수 있게 해주는 것은 물론이고 다양한 함수도 지원하여 사용자가 맞춤형 그래프를 작성할 수 있도록 한다.

### 파이플롯(pyplot)
- 맷플롯립을 이용할 때 가장 기본이 되는 구조는 파이플롯(pyplot)이다.

```python
import matplotlib.pyplot as plt # 맷플롯립 라이브러리 불러오기

X = range(100) # X축 데이터
Y = range(100) # Y축 데이터
plt.plot(X, Y) # 선 그래프 그리기
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1] # 코사인 함수

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2] # 사인 함수

plt.plot(X_1, Y_1) # 코사인 그래프 그리기
plt.plot(X_2, Y_2) # 사인 그래프 그리기
plt.show() # 그래프 출력
```

### 1.2 그림과 축
- 그림(figure)은 그래프를 작성하는 밑바탕이 되는 툴이고, 축(axis)은 일종의 캔버스로 실제 그래프를 작성하는 공간이다.
- subplots 함수를 이용해 하나의 figure에 여러 개의 그래프를 그릴 수 있다.

```python
fig, ax = plt.subplots() # figure와 axis 생성

X_1 = range(100)
Y_1 = [np.cos(value) for value in X] # 코사인 함수

ax.plot(X_1, Y_1) # 코사인 그래프 그리기
ax.set(title='cos gragh',
       xlabel='X', 
       ylabel='Y') # 그래프 제목과 축 이름 설정

plt.show()
```

```python
fig = plt.figure() # figure 생성
fig.set_size_inches(10, 10) # figure 크기 설정

ax_1 = fig.add_subplot(1, 2, 1) # 1행 2열의 첫 번째 subplot
ax_2 = fig.add_subplot(1, 2, 2) # 1행 2열의 두 번째 subplot

ax_1.plot(X_1, Y_1, c="b") # 첫 번째 subplot에 코사인 그래프 그리기
ax_2.plot(X_2, Y_2, c="g") # 두 번째 subplot에 사인 그래프 그리기
plt.show()
```

### 서브플롯 행렬
- 축을 여러 개 만들 땢 일반적으로 figure 객체에서 add_subplot 함수를 사용하거나 plot 객체에서 subplots 함수를 사용하여 축 객체공간을 확보해야 한다.

```python
flg, ax = plt.subplots(nrows = 1, ncols = 2) # 1행 2열의 subplot 생성
print(ax)
```

```python
x = np.linspace(-1, 1, 100) # -1에서 1까지 100개의 균등한 값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2) # 2행 2열의 subplot 생성

ax[0, 0].plot(x, y_1)
ax[0, 1].plot(x, y_2)
ax[1, 0].plot(x, y_3)
ax[1, 1].plot(x, y_4)
```

- matplotlib은 행과 열을 지정한 후, 그 다음 번호는 행과 열이 지정된 행렬에서 위치를 지정한다.

```python
ax1 = plt.subplot(3, 2, 1) # 3행 2열의 첫 번째 subplot
plt.plot(x, y_1)
ax2 = plt.subplot(3, 2, 2) # 3행 2열의 두 번째 subplot
plt.plot(x, y_1)
ax3 = plt.subplot(3, 1, 2) # 3행 1열의 두 번째 subplot
plt.plot(x, y_1)
ax4 = plt.subplot(3, 2, 5) # 3행 2열의 다섯 번째 subplot
plt.plot(x, y_1)
ax5 = plt.subplot(3, 2, 6) # 3행 2열의 여섯 번째 subplot
plt.plot(x, y_1)

plt.show()
```

## 맷플롭립으로 그래프 그리기

### 2.1 색상, 제목
- color 또는 c 매개변수를 사용해서 색상을 변경할 수 있다.
- 16진수의 RGB 값을 입력하는 방법
- 이미 설정된 약어를 사용하는 방법

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color = "#8C58C7")
plt.plot(X_2, Y_2, c="c") # 시안(cyan) 색상

plt.show()
```

```python
plt.plot(X_1, Y_1, color = "#589BC7", linestyle="dashed") # 16진수 색상과 점선
plt.plot(X_2, Y_2, c="y", ls="dotted") # 노란색(yellow) 점선

plt.title("Two lines", fontsize=20) # 그래프 제목
plt.show()
```

```python
# 맷플롯립 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'C:/Users/Admin/AppData/Local/Microsoft/Windows/Fonts/Pretendard-Light.otf' 
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상 해결

fig = plt.figure() # figure 생성
fig.set_size_inches(10, 5) # figure 크기 설정

ax_1 = fig.add_subplot(1, 2, 1) # 1행 2열의 첫 번째 subplot
ax_2 = fig.add_subplot(1, 2, 2) # 1행 2열의 두 번째 subplot

ax_1.plot(X_1, Y_1, c="b") 
ax_1.set_title("그림 1")
ax_2.plot(X_2, Y_2, c="g") 
ax_2.set_title("Figure 2")

plt.show()
```

## 3.2 막대 그래프

```python
data = [[5., 25., 50., 20.], [4., 23., 51., 17.], [6., 22., 52., 19.]]

X = np.arange(0, 8, 2) # X 좌표 시작점 / 0, 2, 4, 6

plt.bar(X + 0.00, data[0], color = 'r', width = 0.50, alpha = 0.5) # 막대 그래프 그리기  
plt.bar(X + 0.50, data[1], color = 'k', width = 0.50, alpha = 0.5)
plt.bar(X + 1.00, data[2], color = 'b', width = 0.50, alpha = 0.5)

plt.xticks(X + 0.50, ("A", "B", "C", "D")) # x축 눈금 설정

plt.show()
```

```python
data = np.array([[5., 25., 50., 20.], [4., 23., 51., 17.], [6., 22., 52., 19.]])

color_list = ['r', 'k', 'b']
data_label = ['A', 'B', 'C']
X = np.arange(data.shape[1])  # X 좌표 시작점 / 0, 1, 2, 3

for i in range(3):
  plt.bar(X, data[i], 
          bottom = np.sum(data[:i], axis=0),
          color = color_list[i],
          label = data_label[i],
          alpha = 0.5) # 누적 막대 그래프 그리기
  print(np.sum(data[:i], axis=0)) # 누적합 / axis=0은 열 방향, axis=1은 행 방향 / [:i]는 0부터 i-1까지의 행 선택 

plt.legend() # 범례 표시
plt.show()
```

## 원 그래프

```python
ratio = [34, 32, 15.5, 18.5]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.01, 0.01, 0.01, 0.2] # 각 조각을 중심에서 얼마나 떨어뜨릴지 설정
colors = ['lightcoral', 'lightskyblue', 'silver', 'gold'] # 각 조각의 색상 설정

# 원 그래프 그리기 / autopct는 각 조각의 비율 표시, startangle은 시작 각도, counterclock는 시계방향 여부, shadow는 그림자 효과
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, colors=colors, shadow=True)
plt.show()
```

## 2 시본에서 사용하는 그래프

## 3 사전 정의된 그래프
- 맷플롯립 관점에서 하나 이상의 그래프들을 합쳐서 다양한 정보를 줄 수 있는 그래프

### 3.1 분포를 나타내는 그래프: 바이올린 플롯과 스웜 플롯
- 바이올린 플롯은 상자그림과 분포도를 한 번에 나타낼 수 있다. violinplot 함수를 사용
= 스웜 플롯은 바이올린 플롯과 같은 형태지만 산점도를 이용해서 데이터의 분포를 나타낸다.

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set2") # 흡연자와 비흡연자의 분포를 나누어 표시
```

```python
sns.swarmplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set2") # 흡연자와 비흡연자의 분포를 나누어 표시
```

### 3.2 그리드 그래프
- FacetGrid 사용

```python
g = sns.FacetGrid(tips, col="time", row="smoker",) # 시간대와 흡연 여부에 따른 그리드 생성
g.map(sns.scatterplot, "total_bill", "tip") # 각 그리드에 산점도 그래프 매핑
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker",) # 시간대와 흡연 여부에 따른 그리드 생성
g.map_dataframe(sns.histplot, "total_bill") # 각 그리드에 산점도 그래프 매핑
# map과 map_dataframe의 차이점은 map은 함수에 데이터프레임 전체를 넘기지 않고, 필요한 열만 넘긴다. 반면에 map_dataframe은 데이터프레임 전체를 넘긴다.
```
