```python
df = pd.read_excel('move.xlsx', header = 0)
df.head(2)
```

```python
df = df.iloc[1:]
df.head(2)
```

```python
df.info()
```

```python
df = df.ffill()
df.head()
```

```python
df.replace('-', 0, inplace = True)
df.head()
```

```python
df_jeju = df[(df['전출지별'] == '제주특별자치도') & (df['전입지별'] != '제주특별자치도')]
df_jeju
```

```python
del df_jeju['전출지별']
df_jeju
```

```python
df_jeju = df_jeju.set_index('전입지별', drop = True)
df_jeju.head()
```

```python
df_jeju = df_jeju.drop("전국", axis = 0)
df_jeju.head()
```

```python
df_jeju_top4 = df_jeju.sort_values(by = '2024', ascending = False).head(4)
df_jeju_top4
```

```python
df_jeju_top4.T.plot(kind = 'area', stacked = False, figsize = (20, 10), alpha = 0.2)
plt.title('제주특별자치도에서 타지역 인구 이동 TOP 4', size = 30)
plt.xlabel('기간')
plt.ylabel('이동자수')
plt.legend(loc = 'best', fontsize = 15)
plt.show()
```
